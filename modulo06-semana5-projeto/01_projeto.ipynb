## Limpeza geral

```python
df_transaction_limpo = df_transaction.copy()
df_transaction_limpo["data"] = pd.to_datetime(df_transaction["DATE"])
df_transaction_limpo["ano"] = df_transaction_limpo["data"].dt.year
df_transaction_limpo["mes"] = df_transaction_limpo["data"].dt.month
df_transaction_limpo["tamanho"] = (
    df_transaction_limpo["PROD_NAME"]
    .str.extract('(\d+)')
    .astype(float)
)
df_transaction_limpo["chip"] = (
    df_transaction_limpo["PROD_NAME"]
    .str.contains("chip", case=False)
)
df_transaction_limpo["marca"] = np.select(
    [
        df_transaction_limpo["PROD_NAME"].str.contains("Cobs", case=False),
        df_transaction_limpo["PROD_NAME"].str.contains("Smiths", case=False),
        df_transaction_limpo["PROD_NAME"].str.contains("Thins", case=False),
        df_transaction_limpo["PROD_NAME"].str.contains("Doritos", case=False),
        df_transaction_limpo["PROD_NAME"].str.contains("Natural Chip", case=False),
        df_transaction_limpo["PROD_NAME"].str.contains("WW", case=False),
        df_transaction_limpo["PROD_NAME"].str.contains("French Fries", case=False)
    ],
    [
        "COBS",
        "SMITHS",
        "THINS",
        "DORITOS",
        "NATURAL CHIP",
        "WW",
        "FRENCH FRIES"
    ],
    "OUTROS"
)
df_transaction_limpo = (
    df_transaction_limpo[df_transaction_limpo["chip"]]
    .reset_index(drop=True)
)
df_transaction_limpo = df_transaction_limpo.drop(columns=["DATE"])
```

```python
df_transaction_limpo
```

## Combinando os dataset´s

```python
df1 = pd.merge(
    df_transaction_limpo,
    df_purchase,
    on="LYLTY_CARD_NBR",
    how="left"
)
```

  - Qual é o preço médio do chip por segmento de cliente

```python
print('Vendas por segmentação: ')
df_vendas_por_segmentacao = (
    df1
    .groupby("PREMIUM_CUSTOMER")
    .agg(
        valor_total_vendas = ("TOT_SALES", "sum"),
        quntidade_produtos_vendidos = ("PROD_QTY", "sum"),
        numero_clientes = ("LYLTY_CARD_NBR", "nunique")
    )
    .reset_index()
)

df_vendas_por_segmentacao
```

```python
df_vendas_por_segmentacao["valor_medio_cliente"] = (
    df_vendas_por_segmentacao["valor_total_vendas"]
    /df_vendas_por_segmentacao["numero_clientes"]
)

df_vendas_por_segmentacao["valor_medio_item"] = (
    df_vendas_por_segmentacao["valor_total_vendas"]
    /df_vendas_por_segmentacao["quntidade_produtos_vendidos"]
)
```

## Vendas por tipo-família

```python
df_vendas_por_tipo_familia = (
    df1
    .groupby("LIFESTAGE")
    .agg(
        valor_total_vendas = ("TOT_SALES", "sum"),
        quntidade_produtos_vendidos = ("PROD_QTY", "sum"),
        numero_clientes = ("LYLTY_CARD_NBR", "nunique")
    )
    .reset_index()
)

df_vendas_por_tipo_familia["valor_medio_cliente"] = (
    df_vendas_por_tipo_familia["valor_total_vendas"]
    /df_vendas_por_tipo_familia["numero_clientes"]
)

df_vendas_por_tipo_familia["valor_medio_item"] = (
    df_vendas_por_tipo_familia["valor_total_vendas"]
    /df_vendas_por_tipo_familia["quntidade_produtos_vendidos"]
)
```

## Vendas por mês e marca

```python
df_vendas_mes_produto = (
    df1
    .groupby(["ano", "mes", "marca"])
    .agg(
        valor_total_vendas = ("TOT_SALES", "sum"),
        quntidade_produtos_vendidos = ("PROD_QTY", "sum"),
        numero_clientes = ("LYLTY_CARD_NBR", "nunique")
    )
    .reset_index()
)

df_vendas_mes_produto["rank_valor_total_vendas"] = (
    df_vendas_mes_produto
    .groupby("marca")
    ["valor_total_vendas"]
    .rank(ascending=False)
)

df_vendas_mes_produto["rank_quntidade_produtos_vendidos"] = (
    df_vendas_mes_produto
    .groupby("marca")
    ["quntidade_produtos_vendidos"]
    .rank(ascending=False)
)
```
