# Análise univariada

A análise univariada tem a função de descrever uma variável isoladamente das demais.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('tips')
```

```python
plt.hist(df.tip);
```

```python
df.tip.hist();
```

```python
sns.histplot(data=df, x='tip');
```

```python
sns.boxplot(data=df, x='total_bill');
```

### Variáveis categóricas

```python
df['sex'].value_counts()
```

```python
(df['sex'].value_counts(normalize=True) * 100).round(2)
```

```python
df['sex'].value_counts().plot(kind='bar');
```

```python
sns.countplot(data=df, x='sex');
```

### Análise bivariada

```python
pd.crosstab(df['sex'], df['smoker'], margins=True)
```

```python
pd.crosstab(df['sex'], df['smoker'], normalize='index')
```

```python
pd.crosstab(df['sex'], df['smoker'], normalize='columns')
```

#### Para facilitar a visualização da dimensão dos números da tabela, podemos utilizar a função “heatmap” do Seaborn, com o parâmetro “annot=True”

```python
sns.heatmap(
    pd.crosstab(df['sex'], df['smoker']),
               annot=True, cmap="Blues"
);
```

### Duas variáveis numéricas

```python
sns.jointplot(data=df,x='total_bill', y='tip');
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex');
```

### Uma variável categórica e uma numérica

```python
df.groupby('sex')['tip'].agg(['mean', 'min', 'max', 'median', 'std'])
```

```python
df.groupby(['sex','smoker'])['tip'].agg(['mean', 'min', 'max', 'median', 'std'])
```

```python
sns.boxplot(data=df, y='tip', x='sex', hue='smoker');
```

## Análise multivariada

consiste em utilizar métodos estatísticos para analisar diversas variáveis. No geral, essas técnicas buscam simplificar ou facilitar a interpretação de um determinado fenômeno a ser estudado com os dados <br>

Análise de componentes principais: <br>

    Do inglês PCA (principal components analysis), é um método de busca construir uma combinação linear das variáveis originais, para representá-las em uma dimensão menor;  
    
Por exemplo: um banco exige que 8 tipos de dados sejam capturados de clientes que buscam empréstimo (renda, idade, tempo na residência atual etc.). Um administrador pode utilizar PCA para reduzir o número de variáveis e facilitar a análise de crédito}

Análise fatorial:

    Seu objetivo é descrever a variabilidade original de um conjunto de variáveis utilizando um número reduzido de variáveis latentes. É muito comum na construção de indicadores;Por exemplo: um gerente de recursos humanos quer determinar quais fatores explicam as 12 variáveis utilizadas para avaliar cada candidato a uma vaga de emprego. Os funcionários classificam cada candidato com base em diversas características, usando uma escala de 1 a 10. A análise fatorial determina quantos fatores respondem pela maior parte do total da variabilidade nos dados.
