```python
df_gap = px.data.gapminder()
df_gap
```

```python
px.scatter(df_gap, x='gdpPercap', y='lifeExp', animation_frame='year', animation_group='country',
          size='pop', color='continent', hover_name='country', facet_col='continent',
          log_x=True, size_max=45, range_x=[100, 1000000], range_y=[25,90])
```

# Bokeh

```python
from bokeh.plotting import figure, output_notebook, show
```

```python
%pip install bokeh
```

```python
# output para notebook
output_notebook()
p = figure(width=400, height=400)

# Adiciona quadrado com tamanho, cor e alpha
p.square([2, 5, 6, 4], [2, 3, 2, 1, 2], size=20, color='navy')

# Mostra os resultados
show(p)
```

```python
# output para notebook
output_notebook()
p = figure(width=400, height=400)

# Adiciona quadrado com tamanho, cor e alpha
p.square([2, 5, 6, 4], [2, 3, 2, 1, 2], size=20, color='navy')
p.line([1, 2, 3, 4, 5], [1, 2, 2, 4, 5], line_width=2) #adicioonou uma linha À figura existente

# Mostra os resultados
show(p)
```

```python
from bokeh.models import HoverTool, BoxSelectTool 

# output para notebook
output_notebook()

# adiciona ferramentas
TOOLS = [BoxSelectTool(), HoverTool()]
p = figure(width=400, height=400, tools=TOOLS)

# Adiciona quadrado com tamanho, cor e alpha
p.square([2, 5, 6, 4, 8], [2, 3, 2, 1, 2], size=20, color='navy', alpha=0.5)

# elementos visuais
p.xaxis.axis_label = 'X-axis'
p.xaxis.axis_label = 'Y-axis'

# Mostra os resultados
show(p)
```
